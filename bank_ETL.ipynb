```python
# Import SQLAlchemy `automap` and other dependencies
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import pandas as pd
```

```python
# Import SQLalchemy data types
from sqlalchemy.types import VARCHAR, INTEGER, DATETIME
```

```python
#Import SQLlite3 to assign Primary Key to table
import sqlite3
```

```python
#Create engine
engine = create_engine("sqlite:///bank_db.db")
conn =engine.connect()
```

```python
#Load bank data
bank_excel =("Resources/banklist_input.xlsx")
bank_df = pd.read_excel(bank_excel, engine='openpyxl')
```

```python
#Create table
bank_table = 'bank_table'
```

```python
#Insert bank data into SQLite bank_db
bank_df.to_sql(bank_table, engine, if_exists='replace', index=False)
```

```python
bank_df
```

```python
#Load Census data
census_excel =("Resources/censuslist_input.xlsx")
census_df = pd.read_excel(census_excel, engine='openpyxl')
```

```python
#Create table
census_table = 'census table'
```

```python
#Insert census data into SQLite bank_db
census_df.to_sql(census_table, engine, if_exists='replace', index=False)
```

```python
census_df
```

```python
#Merge bank_df and census_df
merged_df = pd.merge(bank_df, census_df, how='left', on=['Fund', 'Fund'])
merged_df.head()
```

```python
#Check columns in merged_df
print(merged_df.columns)
```

```python
#rename columns
merged2_df = merged_df.rename(columns={'Bank Name':'Bank_Name','City_x':'City', 'State_x':"State", 'Closing Date_x': 'Bank_Failure_Date', 'Total Housing Units':'Total_Housing_Units','Owner Occupied Units':'Owner_Occupied_Units','Owner Occupied Rate':'Owner_Occupied_Rate', 'Bachelors or higher':'Bachelors_or_higher','Census Year':'Census_Year'})
merged2_df.head(5)
```

```python
#Drop duplict columns and rename columns
merged3_df = merged2_df.drop(['City_y', 'State_y', 'Closing Date_y'], axis=1)

merged3_df.head()
```

```python
#Fill in NaN values from df
[columns_to_fill_1] = ['Total_Housing_Units']
[columns_to_fill_2] = ['Owner_Occupied_Units']
[columns_to_fill_3] = ['Owner_Occupied_Rate']
[columns_to_fill_4] = ['Bachelors_or_higher']
merged3_df[columns_to_fill_1] = merged3_df[columns_to_fill_1].fillna('unavialable')
merged3_df[columns_to_fill_2] = merged3_df[columns_to_fill_2].fillna('unavialable')
merged3_df[columns_to_fill_3] = merged3_df[columns_to_fill_3].fillna('unavialable')
merged3_df[columns_to_fill_4] = merged3_df[columns_to_fill_4].fillna('unavialable')
merged3_df.head()
```

```python
merged3_df.dtypes
```

```python
merged3_df.to_sql('bank_sql', conn, if_exists='replace', index=False)
```

```python
new_merged3_df=pd.read_sql('select * from bank_sql', conn)
new_merged3_df
```

```python
new_merged3_df.columns
```

```python
#Use sqlite3 connection to assign Fund as Primary Key and set data types for columns in 'bank_sql' table
with sqlite3.connect('bank_db.db') as con2:
    merged3_df.to_sql('bank_sql', con=con2, if_exists='replace', index=False, 
                   dtype={'Fund': 'INTEGER PRIMARY KEY', 'Bank_Name':"VARCHAR", 'City':'VARCHAR',
                          'State':'VARCHAR', 'Bank_Failure_Date':'DATETIME', 'Population':'INTEGER',
                          'Total_Housing_Units':'VARCHAR', 'Owner_Occupied_Units':'VARCHAR',
                          'Owner_Occupied_Rate':'VARCHAR','Bachelors_or_higher':'VARCHAR',
                          'Census_Year':'VARCHAR'})
```

```python
#Export merged_df as csv file
merged_df.to_csv("Resources/merged3_df.csv", index=False)
```

```python
#Commit changes and close db connection
engine.dispose()
```
